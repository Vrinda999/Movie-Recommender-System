```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

### Merging Movies and Credits into one CSV File

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
# Removing the Coloumns that are Not Required in our Filtering Process Like: Budget, overview etc.
# Coloumns we are keeping:
#     Genres
#     ID
#     Keywords
#     Title
#     Overview
#     Cast
#     Crew

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

### Removing Missing Data

```python
movies.isnull().sum()
movies.dropna(inplace = True)
```

### Removing Duplicates, if Any.

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# since the genre coloumn is a list of dictionaries, hence we convert it into a list which only stores the genres.

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# Getting a List of the genres of the Movie, instead of string of dictionary format.
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def ConvertforCast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(ConvertforCast)
```

```python
movies.head()
```

```python
def FetchDirector(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(FetchDirector)
```

```python
movies.head()
```

```python
# Converting the Overview into a List.

movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# Removing spaces from keywords, eg: Sam Worthington → SamWorthington

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
# Create a New coloumn called "Tags" which will be the concatenation of all te cols except for ID and Title.

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# Creating a New Dataframe (new_df) which just contains the ID, Title and Tags from the above dataframe.

new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

### Applying Stemming on this Data.
This converts tags like "Dance", "Dancing", "Dances" to "Dance", "Dance", "Dance" i.e. removes similar terms.

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def Stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(Stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

Similarity = Array of Arrays: Similarity of first movies with rest of the movies.

```python
similarity
```

```python
list(enumerate(similarity[0]))           #List of Tuples which shows the Index and it's Similarity Index (To Be Used In the Next Step)
```

## Main Function

```python
def Recommend(Movie):
    Movie_Index = new_df[new_df['title'] == Movie].index[0]                 # To Find out the Index of the Movie so that we can find the Similarity index from Vectors and proceed to recommend movies.
    dists = similarity[Movie_Index]
    MoviesList =  sorted(list(enumerate(dists)), reverse = True, key = lambda x: x[1])[1:6]    # Recommding 5 Movies

    for i in MoviesList:
        print(new_df.iloc[i[0]].title)                 # Prints the Title of the Movies fetched using their Indices
```

## Sample Working

```python
Recommend("Batman Begins")
```
